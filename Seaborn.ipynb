## **a) Plots para Distribuição**

**1) Displot** = Faz histograma dos dados e faz uma curva KDE.


*   Pode tirar a curva sendo kde = False
*   bins =  Altera o tamanho das colunas do histograma, por mais colunas diferentes



```
sns.distplot(tips['total_bill'], kde=False)
```

```
sns.distplot(titanic['fare'], kde=False, color='red')
```

**2) Jointplot** = Cria uma curva semelhante, mas para 2 variaveis ao mesmo tempo.

```
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```
sns.jointplot(x='fare', y='age', data=titanic)
```

**3) Pairplot** = Pegou todas as variaveis numéricas e faz um joinplot de cada uma delas.


*   hue = escolhe uma coluna para mostrar os valores em cores diferentes



```
sns.pairplot(tips, hue='sex')
```

**4) Rugplot** = Cria uma distribuição de cada ponto

```
sns.rugplot(tips['total_bill'])
```

## **b) Distribuição para Dados Categóricos**


*   Dados categoricos são dados que contém texto e poucas possibilidades de valores.



**5) Barplot** = Pede um x e y, especificando o dataset. Pode mostrar um dado estatistico pelo 'estimator'.

```
sns.barplot(x='sex', y='total_bill', data=tips, estimator = np.std)
```

**6) Countplot** = Vai mostrar a contagem do numero de elementos da coluna.

```
sns.countplot(x='sex', data=tips)
```

**7) Boxplot** = 25% dos dados da barra inferior ao começo do quadrado, 50% estão da borda a mediana e 75% da mediana a borda superior. Até a barra superior, mostra a totalidade.



*   hue = Pode acrescentar uma coluna como um boxplot extra para comparar.
*   palette = são as cores
*   orient = 'h', mostra os gráficos na horizontal



```
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex')
```

```
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
```

**8) Violinplot** = Cria distribuições de probabilidades, calcula o kde de cada um e mostra a distribuição dos dados

```
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
```

**9) Stripplot** = Plota pontos aonde os dados aparecem.


*   Separa os dados um pouco, um do lado do outro
*   Split = Quebra as distribuições



```
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

**10) Swarmplot** = Coloca os dados um do lado do outro a medida que aparecem. Nao recomendado para muitos dados

```
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```
sns.swarmplot(x='class', y='age', data=titanic, palette='Set2')
```

**11) Factorplot** = Vira qualquer grafico que atribuir a ele

```
sns.factorplot(x='sex', y='total_bill', data=tips, kind='bar')
```

```
sns.countplot(x='sex', data=titanic)
```

## **c) Plots Matriciais**

```
# Utilizando a base dados fights
flights = sns.load_dataset('flights')
```

```
flights.head()
```

**12) Heatmap** = Mapa de calor.


*   cmap = Cor a ser atribuida a paleta
*   annot = Mostrar os valores, no caso abaixo, das correlações



```
sns.heatmap(tips.corr(), cmap='coolwarm', annot=True)
```

```
# Pivot_table = Controi uma matriz com valores, para comparações
matrix = flights.pivot_table(values='passengers', index='month', columns='year')
matrix
```

```
sns.heatmap(matrix, cmap = 'magma', linecolor='gray', linewidths=1)
```

```
c = titanic.corr()
sns.heatmap(c, cmap='coolwarm')
plt.title=('titanic.corr()')
```

A medida que o tempo passa, a quantidade de voos aumentou e estão concentrados entre Julho/Agosto, no verão deles.

## **d) Plots Regressão**

```
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
```

*   Traça a reta de regressão em um gráfico de dispersão, onde a aréa sobreada de 38 - 50 é a área de maior disperção dos dados.
*   hue = separa os tipos de dados no gráfico
*   markers = mudar o tipo de marcação dos elementos no gráfico
*   col = separa os valores em gráficos, para ficar mais fácil de ver, caso sejam muitos.



```
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='sex')
```

## **e) PairGrids**


*   Constroi gráficos como anteriormente, mas de forma mais poderosa
*   Função map, mapeia todas os gráficos. Pode-se serpara tbm para atingir somente gráficos na diagonal(map_diag(plt.'tipo_grafico')), map_upper(Somente as de cima) e map_lower(de baixo)



```
# New dataset iris
iris = sns.load_dataset('iris')
```

```
iris.head()
```

```
g = sns.PairGrid(iris)
g.map_diag(plt.hist) # Aplica histogramas
g.map(plt.scatter)
```

```
g = sns.PairGrid(iris)
g.map_upper(plt.scatter) # Parte de cima das diagonais do graficos em scatter
```

```
g = sns.PairGrid(iris)
g.map_lower(sns.kdeplot) # Parte de baixo das diagonais do graficos em kde
```

```
sns.pairplot(iris, hue='species') #Grafico nas diagonais em histog. e quadrantes em scatter
```

```
g = sns.FacetGrid(titanic, col="sex")
g.map(plt.hist, "age")
```
